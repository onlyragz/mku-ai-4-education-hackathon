````python
import re

def check_password(password):
    """
    Checks if the password meets criteria and prints the criteria.
      - At least 8 characters long
      - At least one uppercase letter
      - At least one lowercase letter
      - At least one number
      - At least one symbola
    """
    print("Password criteria:")
    print("- At least 8 characters long")
    print("- At least one uppercase letter")
    print("- At least one lowercase letter")
    print("- At least one number")
    print("- At least one symbol")

    # Check for minimum length
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        return False

    # Flags for each criteria
    has_upper = False
    has_lower = False
    has_digit = False
    has_symbol = False

    # Iterate through the password to check for each criteria
    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif re.search("[!@#$%^&*()_+=[\]{};':\"\\|,.<>/?]", char):
            has_symbol = True

    # Check if all criteria are met
    if not has_upper:
        print("Password must contain at least one uppercase letter.")
    if not has_lower:
        print("Password must contain at least one lowercase letter.")
    if not has_digit:
        print("Password must contain at least one number.")
    if not has_symbol:
        print("Password must contain at least one symbol.")

    return has_upper and has_lower and has_digit and has_symbol

def sign_up():
    """
    Prompts user for email, confirms email, and password.
    Checks password and simulates storing user data.
    """
    while True:
        email = input("Enter your email: ")
        confirm_email = input("Confirm your email: ")
        if email!= confirm_email:
            print("Emails do not match. Please try again.")
        else:
            break

    while True:
        password = input("Enter your password: ")
        if check_password(password):
            print("Sign up successful!")
            # In a real application, you would store the email and password securely
            break
        else:
            print("Password does not meet the criteria. Please try again.")

def sign_in():
    """
    Prompts user for email and password, checks password against
    simulated stored data.
    """
    email = input("Enter your email: ")
    password = input("Enter your password: ")
    # In a real application, you would retrieve the stored password for the given email
    # and compare it with the entered password
    if check_password(password):  # This is just a placeholder for actual authentication
        print("Sign in successful!")
    else:
        print("Invalid email or password.")

while True:
    """
    Provides a simple menu for user to choose sign up or sign in.
    """

    choice = input("Do you want to sign up (1) or sign in (2)? ")
    if choice == '1':
        sign_up()
        break
    elif choice == '2':
        sign_in()
        break
    else:
        print("Invalid choice. Please enter 1 or 2.")
````

````python
import time
import threading

# Global variable to keep track of remaining time
remaining_time = 0
is_running = False

def pomodoro_timer_background(study_time, cycles):
    global remaining_time, is_running
    is_running = True
    study_time_seconds = study_time * 60

    for i in range(cycles):
        remaining_time = study_time_seconds
        pomodoro_timer_single_cycle(study_time)
        if i < cycles - 1:  # Don't take a long break after the last cycle
            print("Take a 5-minute short break.")
            time.sleep(300)  # 5 minutes in seconds

    is_running = False
    print("\nAll Pomodoro cycles completed! Take a longer break.")

def pomodoro_timer_single_cycle(study_time):
    global remaining_time
    study_time_seconds = study_time * 60

    start_time = time.time()
    end_time = start_time + study_time_seconds

    while time.time() < end_time:
        remaining_time = int(end_time - time.time())
        minutes = remaining_time // 60
        seconds = remaining_time % 60
        print(f"Time remaining: {minutes:02d}:{seconds:02d}", end="\r")
        time.sleep(1)

    remaining_time = 0  # Timer is done
    print("\nTime's up for this cycle!")

# Get user input
study_time = int(input("Enter study duration (25-120 minutes): "))
cycles = int(input("Enter number of Pomodoro cycles: "))

# Create and start the timer thread
timer_thread = threading.Thread(target=pomodoro_timer_background, args=(study_time, cycles))
timer_thread.start()

print("Timer started in the background. You can continue with other tasks.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import ast

class ProgressTracker:
    def __init__(self, data_file="progress_data.csv"):
        self.data_file = data_file
        try:
            # Check if the file exists and is not empty
            self.df = pd.read_csv(data_file, index_col="Date", parse_dates=True)
        except (FileNotFoundError, pd.errors.EmptyDataError):
            # If the file doesn't exist or is empty, create a new DataFrame
            self.df = pd.DataFrame(columns=["Task", "Completed"])
            self.df['Date'] = pd.to_datetime([])  # Initialize as empty DatetimeIndex
            self.df.set_index('Date', inplace=True)
        except (pd.errors.ParserError, ValueError):
            print("Error reading the CSV file. Creating a new empty DataFrame.")
            self.df = pd.DataFrame(columns=["Task", "Completed"])
            self.df['Date'] = pd.to_datetime([])  # Initialize as empty DatetimeIndex
            self.df.set_index('Date', inplace=True)

    def add_task(self, task, completed=False):
        today = pd.Timestamp(datetime.date.today())  # Use pandas Timestamp
        if today not in self.df.index:
            self.df.loc[today] = [[task], completed]  # Store task as list
        else:
            existing_tasks = self.df.at[today, "Task"]
            if isinstance(existing_tasks, str):
                existing_tasks = ast.literal_eval(existing_tasks)  # Convert string representation back to list
            existing_tasks.append(task)  # Append new task to existing list
            self.df.at[today, "Task"] = existing_tasks  # Update the Task column
        self.df.at[today, "Completed"] = completed  # Set completion status
        self.save_data()

    def mark_complete(self, date, task):
        date_ts = pd.Timestamp(date)
        if date_ts in self.df.index:
            tasks = self.df.at[date_ts, "Task"]
            if isinstance(tasks, list) and task in tasks:
                tasks.remove(task)  # Remove the task if found
                self.df.at[date_ts, "Task"] = tasks  # Update the Task column
                self.df.at[date_ts, "Completed"] = True  # Mark as completed
            else:
                print(f"Task '{task}' not found for date {date}")
        else:
            print(f"Date {date} not found in progress data.")
        self.save_data()

    def calculate_streak(self):
        self.df['Completed'] = self.df['Completed'].astype(bool)
        streaks = self.df['Completed'].groupby((self.df['Completed'] != self.df['Completed'].shift()).cumsum()).transform('size') * self.df['Completed']
        return streaks.iloc[-1] if not streaks.empty else 0  # Handle empty streaks

    def visualize_progress(self):
        daily_completed = self.df["Completed"].astype(int).resample("D").sum()
        streak_data = self.df['Completed'].groupby((self.df['Completed'] != self.df['Completed'].shift()).cumsum()).transform('size') * self.df['Completed']

        fig, ax1 = plt.subplots(figsize=(10, 5))
        ax1.plot(daily_completed.index, daily_completed.values, marker='o', color='blue', label="Daily Completed Tasks")
        ax1.set_xlabel("Date")
        ax1.set_ylabel("Completed Tasks", color='blue')
        ax1.tick_params('y', labelcolor='blue')
        ax2 = ax1.twinx()
        ax2.plot(streak_data.index, streak_data.values, marker='x', color='red', label="Current Streak")
        ax2.set_ylabel("Streak (Days)", color='red')
        ax2.tick_params('y', labelcolor='red')

        plt.title("Daily Task Completion and Streak")
        fig.legend(loc="upper left")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        print(f"Current Streak: {self.calculate_streak()} days")

    def save_data(self):
        # Convert 'Task' column to string representation of lists before saving
        self.df['Task'] = self.df['Task'].apply(lambda x: str(x) if isinstance(x, list) else x)
        self.df.to_csv(self.data_file, index=True)

# Example usage
if __name__ == "__main__":
    tracker = ProgressTracker()  # This will create the CSV if it doesn't exist
    tracker.add_task("Complete homework", completed=False)  # Add a task
    tracker.add_task("Read a book", completed=True)  # Add a completed task
    tracker.visualize_progress()  # Visualize progress
````

````python
class User:
    def __init__(self, username):
        self.username = username
        self.study_time = 0
        self.badges = []
        self.points = 0

class StudySession:
    def __init__(self, room_name, duration):
        self.room_name = room_name
        self.duration = duration
        self.participants = []

    def add_participant(self, user):
        self.participants.append(user)

    def end_session(self):
        for user in self.participants:
            user.study_time += self.duration
            user.points += self.duration * 10  # 10 points per minute
            if self.duration >= 60:  # Award "Focus Master" badge for 60+ minutes
                user.badges.append("Focus Master")

# Leaderboard Logic (simplified):
def update_leaderboard(users):
    return sorted(users, key=lambda user: user.study_time, reverse=True)

# User management
users = []

def add_user(username):
    # Check if user already exists
    if not any(user.username == username for user in users):
        new_user = User(username)
        users.append(new_user)
        print(f"User '{username}' added.")
    else:
        print(f"User '{username}' already exists.")

def start_study_session(username, room_name, duration):
    # Find or create the user
    user = next((u for u in users if u.username == username), None)
    if not user:
        print(f"User '{username}' not found. Please add the user first.")
        return

    session = StudySession(room_name, duration)
    session.add_participant(user)
    session.end_session()  # Simulate ending the session immediately
    print(f"Study session in '{room_name}' ended. User '{username}' now has {user.study_time} study minutes and {user.points} points.")

def display_leaderboard():
    leaderboard = update_leaderboard(users)
    print("\n--- Leaderboard ---")
    for user in leaderboard:
        print(f"Username: {user.username}, Study Time: {user.study_time}, Points: {user.points}, Badges: {', '.join(user.badges)}")
    print("--- End of Leaderboard ---")

# Example usage
if __name__ == "__main__":
    # Add users
    add_user("Alice")
    add_user("Bob")

    # Start study sessions
    start_study_session("Alice", "Room A", 70)  # Alice studies for 70 minutes
    start_study_session("Bob", "Room B", 30)    # Bob studies for 30 minutes

    # Display leaderboard
    display_leaderboard()
````

````python
from flask import Flask, request, jsonify
import numpy as np
import pickle
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity

# Initialize Flask app
app = Flask(__name__)

# Load NLTK resources
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')

# Load the models and embeddings
with open('skill_recommendation_model.pkl', 'rb') as f:
    skill_vectorizer = pickle.load(f)
with open('career_recommendation_model.pkl', 'rb') as f:
    career_vectorizer = pickle.load(f)
with open('glove_embeddings.pkl', 'rb') as f:
    glove_vectors = pickle.load(f)

print("Models and embeddings loaded.")

# Preprocessing function
def preprocess_text(text):
    text = str(text).lower()
    tokens = nltk.word_tokenize(text)
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stopwords.words('english') and token.isalnum()]
    return tokens

# Get embeddings for text tokens
def get_text_embedding(text_tokens):
    embeddings = [glove_vectors[token] for token in text_tokens if token in glove_vectors]
    if embeddings:
        return np.mean(embeddings, axis=0)  # Return the mean of the embeddings
    else:
        return np.zeros(glove_vectors.vector_size)  # Return zero vector if no embeddings found

# Combined Recommendation Route
@app.route('/recommend', methods=['POST'])
def recommend_route():
    data = request.json
    user_interests = data['interests']
    user_course = data['course']
    user_specialization = data['specialization']
    user_skills = data.get('skills', '')  # Optional, for career recommendations
    recommendation_type = data['type']  # "skills" or "careers"

    # Combine input data for embedding
    user_text = f"{user_interests} {user_course} {user_specialization} {user_skills}"
    user_tokens = preprocess_text(user_text)
    user_embedding = get_text_embedding(user_tokens)

    if recommendation_type == "skills":
        # Calculate cosine similarity with skill embeddings
        skillEmbeddings = np.vstack(skill_vectorizer['combined_features_skill_embedding'])
        similarities = cosine_similarity([user_embedding], skillEmbeddings).flatten()
        sorted_indices = similarities.argsort()[-5:][::-1]  # Get top 5 most similar indices
        recommended_skills = [skill_vectorizer['skills'][i] for i in sorted_indices]
        return jsonify({"recommended_skills": recommended_skills})

    elif recommendation_type == "careers":
        # Calculate cosine similarity with career embeddings
        careerEmbeddings = np.vstack(career_vectorizer['combined_features_career_embedding'])
        similarities = cosine_similarity([user_embedding], careerEmbeddings).flatten()
        sorted_indices = similarities.argsort()[-5:][::-1]  # Get top 5 most similar indices
        recommended_careers = [career_vectorizer['first_job_title'][i] for i in sorted_indices]
        return jsonify({"recommended_careers": recommended_careers})

    else:
        return jsonify({"error": "Invalid recommendation type. Use 'skills' or 'careers'."}), 400


````

````python
from flask import Flask, request, render_template, jsonify
import time
import threading

app = Flask(__name__)

# Global variables to keep track of timer status and remaining time
timer_thread = None
remaining_time = 0
is_running = False

def pomodoro_timer_background(study_time, cycles):
    global remaining_time, is_running
    is_running = True
    study_time_seconds = study_time * 60

    for i in range(cycles):
        remaining_time = study_time_seconds
        pomodoro_timer_single_cycle(study_time)
        if i < cycles - 1:  # Don't take a long break after the last cycle
            time.sleep(300)  # 5 minutes in seconds

    is_running = False
    print("\nAll Pomodoro cycles completed! Take a longer break.")

def pomodoro_timer_single_cycle(study_time):
    global remaining_time
    study_time_seconds = study_time * 60

    start_time = time.time()
    end_time = start_time + study_time_seconds

    while time.time() < end_time:
        remaining_time = int(end_time - time.time())
        time.sleep(1)

    remaining_time = 0  # Timer is done

@app.route('/')
def home():
    return render_template('pomodoro.html')

@app.route('/start_timer', methods=['POST'])
def start_timer():
    global timer_thread
    study_time = int(request.form['study_time'])
    cycles = int(request.form['cycles'])

    # Create and start the timer thread
    timer_thread = threading.Thread(target=pomodoro_timer_background, args=(study_time, cycles))
    timer_thread.start()

    return "Timer started! You can continue with other tasks."

@app.route('/time_remaining', methods=['GET'])
def time_remaining():
    return jsonify({
        'remaining_time': remaining_time,
        'is_running': is_running
    })


````

````python
from flask import Flask, request, jsonify, render_template
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import ast
import os

app = Flask(__name__)
data_file = "progress_data.csv"  # Define your data file here

class ProgressTracker:
    def __init__(self, data_file=data_file):
        self.data_file = data_file
        try:
            self.df = pd.read_csv(data_file, index_col="Date", parse_dates=True)
        except (FileNotFoundError, pd.errors.EmptyDataError):
            self.df = pd.DataFrame(columns=["Task", "Completed"])
            self.df['Date'] = pd.to_datetime([])  # Initialize as empty DatetimeIndex
            self.df.set_index('Date', inplace=True)
        except (pd.errors.ParserError, ValueError):
            self.df = pd.DataFrame(columns=["Task", "Completed"])
            self.df['Date'] = pd.to_datetime([])  # Initialize as empty DatetimeIndex
            self.df.set_index('Date', inplace=True)

    def add_task(self, task, completed=False):
        today = pd.Timestamp(datetime.date.today())
        if today not in self.df.index:
            self.df.loc[today] = [[task], completed]
        else:
            existing_tasks = self.df.at[today, "Task"]
            if isinstance(existing_tasks, str):
                existing_tasks = ast.literal_eval(existing_tasks)  # Convert string back to list
            existing_tasks.append(task)
            self.df.at[today, "Task"] = existing_tasks
        self.df.at[today, "Completed"] = completed
        self.save_data()

    def mark_complete(self, date, task):
        date_ts = pd.Timestamp(date)
        if date_ts in self.df.index:
            tasks = self.df.at[date_ts, "Task"]
            if isinstance(tasks, list) and task in tasks:
                tasks.remove(task)
                self.df.at[date_ts, "Task"] = tasks
                self.df.at[date_ts, "Completed"] = True
            else:
                return f"Task '{task}' not found for date {date}"
        else:
            return f"Date {date} not found in progress data."
        self.save_data()

    def visualize_progress(self):
        daily_completed = self.df["Completed"].astype(int).resample("D").sum()
        streak_data = self.df['Completed'].groupby((self.df['Completed'] != self.df['Completed'].shift()).cumsum()).transform('size') * self.df['Completed']

        plt.figure(figsize=(10, 5))
        plt.plot(daily_completed.index, daily_completed.values, marker='o', label="Daily Completed Tasks", color='blue')
        plt.title("Daily Task Completion")
        plt.xlabel("Date")
        plt.ylabel("Completed Tasks")
        plt.legend()
        plt.grid()
        plt.savefig("progress_plot.png")  # Save the plot as an image
        plt.close()

    def save_data(self):
        self.df['Task'] = self.df['Task'].apply(lambda x: str(x) if isinstance(x, list) else x)
        self.df.to_csv(self.data_file, index=True)

# Create a global ProgressTracker instance
tracker = ProgressTracker()

@app.route('/')
def index():
    return render_template('index.html')  # Create an HTML file to serve

@app.route('/add_task', methods=['POST'])
def add_task():
    data = request.json
    task = data.get("task")
    completed = data.get("completed", False)
    tracker.add_task(task, completed)
    return jsonify({"message": "Task added successfully!"}), 201

@app.route('/mark_complete', methods=['POST'])
def mark_complete():
    data = request.json
    date = data.get("date")
    task = data.get("task")
    message = tracker.mark_complete(date, task)
    if message:
        return jsonify({"message": message}), 400
    return jsonify({"message": "Task marked as complete!"}), 200

@app.route('/visualize_progress', methods=['GET'])
def visualize_progress():
    tracker.visualize_progress()  # Generates and saves the progress plot
    return jsonify({"message": "Progress visualization created!"}), 200

@app.route('/current_streak', methods=['GET'])
def current_streak():
    streak = tracker.calculate_streak()
    return jsonify({"current_streak": streak}), 200

````

````python
from flask import Flask, request, jsonify

app = Flask(__name__)

class User:
    def __init__(self, username):
        self.username = username
        self.study_time = 0
        self.badges = []
        self.points = 0

class StudySession:
    def __init__(self, room_name, duration):
        self.room_name = room_name
        self.duration = duration
        self.participants = []

    def add_participant(self, user):
        self.participants.append(user)

    def end_session(self):
        for user in self.participants:
            user.study_time += self.duration
            user.points += self.duration * 10  # 10 points per minute
            if self.duration >= 60:  # Award "Focus Master" badge for 60+ minutes
                user.badges.append("Focus Master")

# Leaderboard Logic (simplified):
def update_leaderboard(users):
    return sorted(users, key=lambda user: user.study_time, reverse=True)

# User management
users = []

@app.route('/add_user', methods=['POST'])
def add_user():
    username = request.json.get('username')
    if not username:
        return jsonify({"error": "Username is required"}), 400

    if any(user.username == username for user in users):
        return jsonify({"message": f"User '{username}' already exists."}), 400

    new_user = User(username)
    users.append(new_user)
    return jsonify({"message": f"User '{username}' added."}), 201

@app.route('/start_session', methods=['POST'])
def start_session():
    username = request.json.get('username')
    room_name = request.json.get('room_name')
    duration = request.json.get('duration')

    if not username or not room_name or duration is None:
        return jsonify({"error": "Username, room name, and duration are required"}), 400

    user = next((u for u in users if u.username == username), None)
    if not user:
        return jsonify({"error": f"User '{username}' not found. Please add the user first."}), 404

    session = StudySession(room_name, duration)
    session.add_participant(user)
    session.end_session()  # Simulate ending the session immediately

    return jsonify({
        "message": f"Study session in '{room_name}' ended.",
        "study_time": user.study_time,
        "points": user.points
    }), 200

@app.route('/leaderboard', methods=['GET'])
def leaderboard():
    leaderboard = update_leaderboard(users)
    result = [
        {
            "username": user.username,
            "study_time": user.study_time,
            "points": user.points,
            "badges": user.badges
        }
        for user in leaderboard
    ]
    return jsonify(result), 200


````
